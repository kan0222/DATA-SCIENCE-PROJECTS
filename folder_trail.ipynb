```python
import matplotlib.pyplot as plt
import pandas as pd

allan = pd.read_csv('/content/BMW sales data (2010-2024) (1).csv')

# Count models
model_counts = allan["Region"].value_counts()

# Pie chart
plt.pie(
    model_counts,
    labels=model_counts.index,
    autopct='%1.1f%%',
    shadow=True,
    startangle=140
)
plt.title("BMW Sales Data (2010â€“2024) by Region")
plt.show()
```
